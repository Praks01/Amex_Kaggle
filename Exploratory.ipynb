```python
df_train['D_63'] = df_train['D_63'].astype('category')
df_train['D_64'] = df_train['D_64'].astype('category')
```

```python
enc=OneHotEncoder()
enc_data=pd.DataFrame(enc.fit_transform(df_train[['D_63_new','D_64_new']]).toarray())
```

```python
df_train = df_train.join(enc_data)
```
